```python
dataframe = pd.read_csv('advertising_Classification - advertising_Classification.csv')

dataframe
```

```python
X = dataframe[['Daily Time Spent on Site', 'Area Income','Daily Internet Usage','Age']]
y = dataframe['Clicked on Ad']

X_test , X_train , y_test, y_train = train_test_split(X,y, test_size = 0.20, random_state= 42)

logistics = LogisticRegression()

feedings= logistics.fit(X_train,y_train)

prediction = feedings.predict(X_test)
score = accuracy_score(y_test, feedings.predict(X_test))

print("Prediction : " , prediction)
print("Accuracy Score : ", score)




```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, prediction)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Clicked On ads", " Not Clicked On Ads"], yticklabels=["Clicked On ad", "Not Clicked On Ads"])
plt.title("Confusion Matrix (Logistics Regression)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
