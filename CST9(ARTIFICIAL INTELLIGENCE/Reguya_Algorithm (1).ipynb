```python

read_csv = pd.read_csv('heart-disease-cleveland.csv')

print(read_csv)
```

```python
read_csv.columns = read_csv.columns.str.strip()


read_csv[['ca', 'thal']] = read_csv[['ca', 'thal']].replace('?', pd.NA)


x = read_csv[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']].apply(pd.to_numeric, errors='coerce')
y = read_csv['diagnosis']


x = x.fillna(x.mean())

```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
algorithm = GaussianNB()

train = algorithm.fit(x_train,y_train)
pred = algorithm.predict(x_test)
score = accuracy_score(y_test,pred)


```
