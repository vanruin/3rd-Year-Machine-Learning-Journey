```python
import pandas as pd
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

filez = files.upload()
```

```python
filezx = pd.read_csv('covid_toy - covid_toy (1).csv')
filezx
```

```python
x = filezx[['age','fever']]

y = filezx['has_covid']

predictx = DecisionTreeClassifier()

x_train, x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)



wow = predictx.fit(x_train,y_train)
prdc = predictx.predict(x_test)

score = accuracy_score(y_test,prdc)
print(wow)
print(prdc)
print(score)
```
