```python
import pandas as pd
read = pd.read_csv('patient_priorityRev1 - patient_priorityRev1 (1).csv')

#Load the dataset and display the first 10 rows.
read.head(10)
```

```python
#Check all for missing values. If any, handle them appropriately. for gender , age , hypertension, triage, residence_type
print("Before : ", read.isnull().sum())
handling = read['heart_disease'].mean()
read['heart_disease'] = read['heart_disease'].fillna(handling)
read.dropna(subset=['gender', 'age', 'hypertension', 'triage', 'Residence_type'], inplace=True)
print("\nAfter : ",read.isnull().sum())

```

```python
#(Normalize numerical variables (age, blood pressure, cholesterol, max heart rate)
from sklearn.preprocessing import StandardScaler

numerical_cols = ['age','blood pressure','cholesterol','max heart rate']
scaler = StandardScaler()
read[numerical_cols] = scaler.fit_transform(read[numerical_cols])


```

```python
#Split the dataset into 80% training and 20% testing.
from sklearn.model_selection import train_test_split


x = read.drop('triage', axis=1)
y = read['triage']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Shape of x_train:", x_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```
